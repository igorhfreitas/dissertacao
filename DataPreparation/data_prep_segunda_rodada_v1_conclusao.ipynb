```python
df.shape
```

```python
df=df.drop(df.filter(regex="x\d+$|x\d+_RF2$|y\d+$|y\d+_RF2$").columns,axis=1)
```

```python
for col in df.filter(regex="x|y").columns:
    df[col]=df[col].str.replace("[","").str.replace("]","").str.replace("(","").str.replace(")","").str.strip().astype("float")
    df=df.rename(columns={col:col.replace("_RF1","")})
df
```

```python
df["Key"]=(df.reset_index()["index"].astype(str)+"_"+df["simul_time"].astype(str)+"_"+df["material"].astype(str)+"_"+df["depth"].astype(str)+"_"
            +df["vc"].astype(str)+"_"+df["rake_angle"].astype(str)+"_"+df["clearance_angle"].astype(str))
df_pivoted,temp_df=pd.DataFrame(),pd.DataFrame()
for i in df.itertuples():
    key=i[-1]
    temp_df_x=df.iloc[[i[0]]].filter(regex="x\d+").T.rename(columns={i[0]:"x"}).reset_index(drop=True)
    temp_df_y=df.iloc[[i[0]]].filter(regex="y\d+").T.rename(columns={i[0]:"y"}).reset_index(drop=True)
    temp_df=pd.concat([temp_df_x,temp_df_y],axis=1).assign(Key=key)
    if df_pivoted.empty:
        df_pivoted=temp_df
    else:
        df_pivoted=pd.concat([df_pivoted,temp_df])
df_pivoted.head()
```

```python
df_pivoted_2=df_pivoted.merge(df[["Key","simul_time","material","depth","vc","rake_angle","clearance_angle"]],
                              on="Key",
                              how="left")
df_pivoted_2
```

```python
from scipy import stats
df.filter(regex="^y")
map_outlier_df=pd.pivot_table(df.filter(regex="y"),
                index="Key",
                )
mask_=map_outlier_df.apply(stats.zscore,axis=1).apply(lambda x: x<0.25)
cleaned_df=pd.pivot_table(df.filter(regex="y"),
                index="Key",
                ).mask(mask_)

cleaned_df=cleaned_df[["y1","y2","y3","y4","y5","y6","y7","y8","y9","y10","y11",
                    "y12","y13","y14","y15","y16","y17","y18","y19","y20","y21"]]

def mean_last_5_highest_non_null(row):
    sorted_values = row.dropna().sort_values(ascending=False)  # Drop NaN and sort
    return sorted_values.head(5).mean()

cleaned_df["expected_cut_force"] = cleaned_df.filter(regex="y").apply(mean_last_5_highest_non_null, axis=1)
```
